### Init function

```python
def make_dict(row):
    context_dct = {}
    token_lst = row['contexts'].replace("'",'"').split()
    context_dct['verbs'] = []
    # context_tags_list = ast.literal_eval(row['context_tags'])
    for context_tags_iter in row['context_tags']:
        context_tags = ast.literal_eval(context_tags_iter)
        indices = [i for i, x in enumerate(context_tags) if x == "B-PREDICATE-ACT"]
        for i in indices:
            verb = {}
            verb["verb"] = token_lst[i]
            verb["description"] = create_descriptions(token_lst, context_tags)
            verb["tags"] = context_tags
            context_dct["verbs"].append(verb)
    context_dct['words'] = token_lst    
    return context_dct
```

```python
def create_descriptions(token, tags):
    s = ''
    for token, tag in zip(token, tags):
        if tag == 'O':
            s += token + ' '
        else:
            s += f'[{tag}: {token}] '
    return s
```

```python
def start_end_index(example):
    tok_to_orig_index = []
    orig_to_tok_index = []
    all_doc_tokens = []
    for i, token in enumerate(example['contexts'].split()):
        for t in range(len(token)+1):
            tok_to_orig_index.append(i)
        all_doc_tokens.append(token)
    if example['labels']:
        start_index = -1
        end_index = -1
    else:
        start_index = tok_to_orig_index[int(example['start'])]
        end_index = tok_to_orig_index[int(example['end'])]
    return start_index, end_index
```
