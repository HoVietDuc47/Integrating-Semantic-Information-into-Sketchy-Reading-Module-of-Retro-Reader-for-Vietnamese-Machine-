```python
import pandas as pd
import ast

df = pd.read_csv("/content/drive/MyDrive/KLTN/srl/data_7k/data_prediction_2.csv")
df['label'] = pd.read_csv("/content/drive/MyDrive/KLTN/srl/data_original/data_train_1.csv")['is_impossible'][:7000]
df.head(3)
```

```python
i_lst = []
for i in range(len(df['context_tags'])):
    if 'B-PREDICATE-ACT' not in df['context_tags'][i] or 'B-PREDICATE-ACT' not in df['question_tags'][i]:
        i_lst.append(i)
df_drop = df.drop(labels=i_lst, axis=0).reset_index(drop=True)
```

```python
def create_descriptions(token, tags):
    s = ''
    for token, tag in zip(token, tags):
        if tag == 'O':
            s += token + ' '
        else:
            s += f'[{tag}: {token}] '
    return s
```

```python
def make_dict(row):
    context_dct = {}
    token_lst = row['contexts'].split()
    context_tags = ast.literal_eval(row['context_tags'])
    context_dct['verbs'] = []
    indices = [i for i, x in enumerate(context_tags) if x == "B-PREDICATE-ACT"]
    for i in indices:
        verb = {}
        verb["verb"] = token_lst[i]
        verb["description"] = create_descriptions(token_lst, context_tags)
        verb["tag"] = context_tags
        context_dct["verbs"].append(verb)
    context_dct['words'] = token_lst
    return context_dct
```

```python
def make_dict_question(row):
    context_dct = {}
    token_lst = row['questions'].split()
    context_tags = ast.literal_eval(row['question_tags'])
    context_dct["verbs"] = []
    indices = [i for i, x in enumerate(context_tags) if x == "B-PREDICATE-ACT"]
    for i in indices:
        verb = {}
        verb["verb"] = token_lst[i]
        verb["description"] = create_descriptions(token_lst, context_tags)
        verb["tag"] = context_tags
        context_dct["verbs"].append(verb)
    context_dct["words"] = token_lst
    
    return context_dct
```

```python
context_dict_list = [make_dict(df_drop.iloc[i]) for i in range(len(df_drop))]
question_dict_list = [make_dict_question(df_drop.iloc[i]) for i in range(len(df_drop))]
```

```python
df_final = pd.DataFrame({
    "index": range(len(df_drop)), 
    "sentence1": df_drop['contexts'],
    "sentence2": df_drop['questions'],
    "{'verbs': [], 'words': ['sentence1']}": context_dict_list,
    "{'verbs': [], 'words': ['sentence2']}": question_dict_list,
    'gold_label': df_drop['label']
    })
```

```python
df_final.to_csv("/content/drive/MyDrive/KLTN/srl/data_7k_sembert/data_2.tsv_tag_label", sep="\t", index=False)
```
