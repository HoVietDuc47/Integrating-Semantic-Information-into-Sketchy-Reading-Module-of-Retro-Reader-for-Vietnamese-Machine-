## Convert to CSV

```python
import json
import pandas as pd

json_data = json.load(open("/content/drive/MyDrive/KLTN/data/kltn/train.json"))['data']
json_data_dev = json.load(open("/content/drive/MyDrive/KLTN/data/kltn/dev.json"))['data']
json_data_test = json.load(open("/content/drive/MyDrive/KLTN/data/kltn/test.json"))['data']
```

```python
def convert_csv(json_data): 
    contexts, questions, is_impossible, answers, qa_id = [], [], [], [], []
    pr = 0
    answers_start, answers_end = [], []
    for data in json_data:
        for para in data['paragraphs']:
            
            for qa in para['qas']:
                contexts.append(para['context'])
                questions.append(qa['question'])
                is_impossible.append(qa['is_impossible'])
                qa_id.append(qa['id'])
                pr = pr + 1
                if qa['is_impossible'] == False:

                    answers.append(qa['answers'][0]['text'])
                    answers_start.append(qa['answers'][0]['answer_start'])
                    answers_end.append(qa['answers'][0]['answer_start']+len(qa['answers'][0]['text']))
                else:
                    
                    answers.append("")
                    answers_start.append("")
                    answers_end.append("")
    df = pd.DataFrame({"id":qa_id, "contexts":contexts,"questions":questions, "is_impossible":is_impossible, "answers":answers, "start":answers_start,"end":answers_end})
    print(pr)
    return df
```

```python
df = convert_csv(json_data)
df_dev = convert_csv(json_data_dev)
df_test = convert_csv(json_data_test)
```
